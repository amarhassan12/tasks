```python
import pandas as pd
import matplotlib.pyplot as plt

car_sales = pd.DataFrame({
    "Odometer (KM)": [5000, 15000, 23000, 45000, 60000, 72000, 88000, 102000, 125000, 150000],
    "Price": ["10000", "15000", "20000", "25000", "30000", "35000", "40000", "45000", "50000", "55000"]
})

car_sales["Price"] = car_sales["Price"].astype(int)

car_sales.plot(x="Odometer (KM)", y="Price", kind="scatter")
plt.show()
```
